```python
###Creating the new col
```

```python
cond = ((df['DIED'] == 'Y') | (df['ER_VISIT'] == 'Y') | (df['HOSPITAL'] == 'Y') | (df['DISABLE'] == 'Y'))
```

```python
df['SERIOUS'] = np.where(cond, 'Y', 'N')
```

```python
df['SERIOUS'].value_counts()
```
